```python
from crewai import Agent, Task, Crew, LLM
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
# Initialize the Ollama LLM
llm = LLM(model="ollama/llama3.1", base_url="http://localhost:11434")
```

```python
# Define your agents
researcher = Agent(
    role="Researcher",
    goal="Discover new insights on AI trends",
    backstory="An AI researcher focused on the latest developments in artificial intelligence.",
    llm=llm,
    verbose=True,
)

writer = Agent(
    role="Writer",
    goal="Create an engaging article on AI advancements",
    backstory="A technical writer specializing in AI and machine learning topics.",
    llm=llm,
    verbose=True,
)
```

```python
# Define tasks for the agents
task1 = Task(
    description="Investigate the latest AI trends and compile key findings.",
    expected_output="A summary of current AI trends.",
    agent=researcher,
)

task2 = Task(
    description="Write a comprehensive article based on the research findings.",
    expected_output="An article draft on AI advancements.",
    agent=writer,
)
```

```python
# Create the crew and execute tasks
crew = Crew(agents=[researcher, writer], tasks=[task1, task2], verbose=True)
```

```python
# Start the process
result = crew.kickoff()
print(result)
```
